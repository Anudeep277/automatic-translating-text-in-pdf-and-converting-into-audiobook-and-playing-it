```
#Translating text in pdf from one language to another and making it into audiobook and playing it AUTOMATICALLY
```

```
import PyPDF2
from gtts import gTTS
from IPython.display import Audio
import os
```

```
pdfFileObj = open('/content/summary.pdf', 'rb')
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
num_pages = pdfReader.numPages
for num in range(2,num_pages): 
  pageObj = pdfReader.getPage(num)
result=pageObj.extractText()
translated = GoogleTranslator(source='auto', target='te').translate(result)
print(translated)
f = open("telugu.txt", "a")
f.write(translated)
f.close()
file=open("telugu.txt","r",encoding='utf8').read().replace("\n"," ")
speech=gTTS(text=str(file),lang="te",slow=False)
speech.save("telugu.mp3")
Audio("telugu.mp3",autoplay=True)

```
